```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

import warnings
warnings.filterwarnings('ignore')

x = np.linspace(0, 2*np.pi, 200)
y = np.sin(x)

#set figure size
plt.figure(figsize=(10, 10))

sns.lineplot(x=x, y=y, label='sin(x)')

sumfreq = 0

for i in range(1, 5):
    fy = np.fft.fft(y)
    
    fyreal = fy.real
    fyimag = fy.imag
    
    fyreal = np.roll(fyreal, -i)
    fyimag = np.roll(fyimag, i)
    
    fy = i*fyreal + 1j * fyimag
    
    z = np.fft.ifft(fy)
    
    sumfreq += z
    
    sns.lineplot(x=x, y=z, label=f'i={i}')

sns.lineplot(x=x, y=sumfreq, label='sumfreq')
plt.show()

```
